```python
BEKE_data.rename(columns={'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close', 'Adj Close': 'adj_close', 'Volume': 'volume'}, inplace=True)
```

```python
cerebro = bt.Cerebro()
```

```python
class A_Strategy(bt.Strategy):
    pass
```

```python
data = bt.feeds.PandasData(dataname = BEKE_data)
```
